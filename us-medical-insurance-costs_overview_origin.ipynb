# U.S. Medical Insurance Costs

```python
import csv
insurance_data = {}
with open("insurance.csv") as insurance_csv:
    insurance_data = csv.DictReader(insurance_csv)
    age_list=[]
    sex_list = []
    bmi_list = []
    children_list = []
    smoker_list = []
    region_list = []
    charges_list = []
    for row in insurance_data:
        age_list.append(row["age"])
        sex_list.append(row["sex"])
        bmi_list.append(row["bmi"])
        children_list.append(row["children"])
        smoker_list.append(row["smoker"])
        region_list.append(row["region"])
        charges_list.append(row["charges"])
    insurance_data={"age":age_list, "sex": sex_list, "bmi": bmi_list, "children": children_list, "smoker": smoker_list, "region": region_list, "charges": charges_list}
print(insurance_data["children"])


#daten√ºberblick: mittleres alter, geschlecht, kosten, kinder, raucher
#{'age' 'sex':  'bmi': 'children': 'smoker': 'region', 'charges'}
# library mit key ages, sex, bmi, children, smoker, region, charges
# library mit key nr der person und allen werten als valuables
```

```python
##overview of averages
total_nr= len(insurance_data["age"])
print("There is a total of " + str(total_nr) + " people involved.")

#age
total_age = 0
for age in insurance_data["age"]:
    total_age += int(age) 
    avg_age= total_age/len(insurance_data["age"])
total_age = 0
for age in insurance_data["age"]:
    total_age += int(age)
avg_age= total_age/len(insurance_data["age"])
print("The average age is " + str(round(avg_age,2)) + " years.")

#smokers
total_smokers= 0
total_non_smokers = 0
for smokers in insurance_data["smoker"]:
    if smokers == "yes":
        total_smokers+=1
    if smokers == "no":
        total_non_smokers+=1
check_if_im_dumb = total_smokers + total_non_smokers
if check_if_im_dumb == total_nr:
    print("Good job")
else:
    print("uh-oh")
print("There are "+ str(total_smokers) + " smokers and " + str(total_non_smokers) + " non-smokers, which adds up to a total of " + str(check_if_im_dumb) + ".") 

#sex
males= 0
females = 0
for sex in insurance_data["sex"]:
    if sex == "male":
        males+=1
    if sex == "female":
        females+=1
check_if_im_dumb = females + males
if check_if_im_dumb == total_nr:
    print("Good job")
else:
    print("uh-oh")
print("There are "+ str(males) + " females and " + str(males) + " males, which adds up to a total of " + str(check_if_im_dumb) + ".") 

#children
parents=0
for children in insurance_data["children"]:
    if int(children) > 0:
        parents+=1
        parents_percent= round(parents/total_nr*100,1)
print("There are "+ str(parents) + " parents, equal to " + str(parents_percent) + "%.")

#cost
total_charges=0
for charges in insurance_data["charges"]:
    total_charges += float(charges)
charges_per_person= total_charges/total_nr
print("The average cost is "+ str(round(charges_per_person,2)) + "$ per person.")

#regions
list_of_regions = []
people_per_region={}
for region in insurance_data["region"]:
    if region not in list_of_regions:
        list_of_regions.append(region)
for region in insurance_data["region"]:
    if region not in people_per_region.keys():
        people_per_region[region] = 1

    elif region in people_per_region.keys():
        people_per_region[region]+=1
for region in people_per_region:
    percent_per_region = round(int(people_per_region[region])/total_nr*100,2)
    people_per_region[region] = [people_per_region[region], percent_per_region]
print(people_per_region)   

#bmi
total_bmi = 0.0
for bmi in insurance_data["bmi"]:
    total_bmi+= float(bmi)
avg_bmi= total_bmi/total_nr
print("The average BMI is " + str(round(avg_bmi,2)) + ", rounded.")
    

#overview of medians for age, cost and bmi
#bmi
bmi_sorted = sorted(insurance_data["bmi"]) 
bmi_median = (float(bmi_sorted[int(total_nr/2)-1]) + float(bmi_sorted[int(total_nr/2)]))/2
print("The median BMI is " + str(bmi_median) + ".")

#age
age_sorted = sorted(insurance_data["age"])
age_median = (float(age_sorted[int(total_nr/2)-1]) + float(age_sorted[int(total_nr/2)]))/2
print("The median age is " + str(age_median) + ".")

#charges
charges_sorted = sorted(insurance_data["charges"]) 
charges_median = (float(charges_sorted[int(total_nr/2)-1]) + float(charges_sorted[int(total_nr/2)]))/2
print("The median cost is " + str(round(charges_median,2)) + ".")
```

```python
def find_total_nr():
    total_nr= len(insurance_data["age"])
    return total_nr

def find_avg_age():
    total_age = 0
    for age in insurance_data["age"]:
        total_age += int(age)
        
    avg_age= total_age/len(insurance_data["age"])
    print("The average age is " + str(avg_age) + " years.")
    return avg_age
 
def find_total_smokers():
    total_smokers= 0
    total_non_smokers = 0
    for smokers in insurance_data["smoker"]:
        if smokers == "yes":
            total_smokers+=1
        if smokers == "no":
            total_non_smokers+=1
    check_if_im_dumb = total_smokers + total_non_smokers
    if check_if_im_dumb == find_total_nr():
        print("Good job")
    else:
        print("uh-oh")
    print("There are "+ str(total_smokers) + " smokers and " + str(total_non_smokers) + " non-smokers, which adds up to a total of " + str(check_if_im_dumb) + ".") 
    return total_smokers, total_non_smokers

def find_total_by_sex():
    males= 0
    females = 0
    for sex in insurance_data["sex"]:
        if sex == "male":
            males+=1
        if sex == "female":
            females+=1
    check_if_im_dumb = females + males
    if check_if_im_dumb == find_total_nr():
        print("Good job")
    else:
        print("uh-oh")
    print("There are "+ str(males) + " females and " + str(males) + " males, which adds up to a total of " + str(check_if_im_dumb) + ".") 
    return females, males

def find_nr_of_parents():
    parents=0
    for children in insurance_data["children"]:
        if int(children) > 0:
            parents+=1
            parents_percent= round(parents/find_total_nr()*100,1)
    print("There are "+ str(parents) + " parents, equal to " + str(parents_percent) + "%.")
    return parents

def find_average_cost():
    total_charges=0
    for charges in insurance_data["charges"]:
        total_charges += float(charges)
    charges_per_person= total_charges/total_nr
    print("The average cost is " + "$ per person.")
    return charges_per_person
def get_an_overview():
    find_avg_age()
    find_total_nr()
    find_total_smokers()
    find_total_by_sex()
    find_nr_of_parents()
    find_average_cost()
get_an_overview()
```
